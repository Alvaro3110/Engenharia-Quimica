a) Desenvolva uma função estruturada para o cálculo da altura do nível do tanque dado um intervalo de simuação. Em seguida, crie um gráfico da altura versus tempo. Utilize um tempo de simulação de ti = 0 a tf = 20 min com um passo de tempo Delta = 1s. Em qual valor a altura do nível se estabiliza ?

b) De acordo com a solução analítica, a altura é proporcional à vazão de alimentação do tanque. Assim, se Fi cai pela metade, a altura também deverá cair. Mostre que isso é verdade construindo o gráfico altura versus tempo para Fi = 4m^3/min



```octave
clc;
clear all;

variacao_do_tempo = 1/60; #1s dt
t_inicial = 0;
tempo_final = 20; #min tf


r = 0.125; #min/m2
a = 0.3; # m2
Fi =  4; #m3/min
tempo = t_inicial:variacao_do_tempo:tempo_final;

for i = 1:length(tempo)
  altura(i)= r*Fi*(1- exp((-tempo(i))/(r*a)));
 end

 plot(tempo,altura);
 xlabel('Tempo(min)');
 ylabel('Altura (m)');
 title('Altura vs. Tempo');
 plot(tempo,altura);
 set(gcf,'Visible','on') #Este codigo é apenas para integrar o octave ao jupyter notebook sendo ele necessario para plotar o gráfico
```
