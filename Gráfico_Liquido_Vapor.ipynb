```
import numpy as np
import matplotlib.pyplot as plt

NC = 2
T = 266.6 # É aqui que vc quer encontrar
xx = np.arange(0,1,0.01)

A = np.array([9.058,8.9179])
B = np.array([2154.9,2032.76])
C = np.array([-34.42,-33.15])


Psat = np.zeros(NC)
x = np.zeros(NC)
P = np.zeros(xx.size)
y = np.zeros((xx.size,NC))

for i in range(NC):
  Psat[i] = np.exp(A[i]-B[i]/(T+C[i]))
i = 0
for x1 in xx:
  x = np.array([x1,1-x1])
  P[i] = np.sum(np.dot(x,Psat))
  for j in range(NC):
    y[i,j] = x[j]*Psat[j]/P[i]
  i += 1 

plt.plot(xx,P,label ='fração molar liquida $x_1$')
plt.plot(y[:,0],P,label='fração molar vapor $y_1$')
plt.legend(loc='best')
plt.xlabel('$x_1$,$y_1$')
plt.ylabel('Pressão [bar]')
plt.grid(color='k',linestyle=':')


for i in range(0,len(A)):
  print("Valor de A",[i],": ", A[i])
for v in range(0,len(B)):
  print("Valor de B",[v],": ", B[v])
```
